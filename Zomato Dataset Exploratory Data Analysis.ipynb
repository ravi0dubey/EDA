```python
#another way of finding specific columns which has null values

list_null_features= []

for features in df_zomato.columns:
    if df_zomato[features].isnull().sum()>0:
        list_null_features.append(features)   
        
print(list_null_features)
```

```python
#Using heatmap to plot missing values
sns.set(rc = {'figure.figsize':(10,20)})
sns.heatmap(df_zomato.isnull(),yticklabels=False,cbar= False,cmap = 'viridis')
```

```python
# To Arrange above data into Dataframe we need to convert it using reset_index
# Count does not have any Header label, Header Label will get displayed as 0
# so we will use rename function to set its name to Rating_Count
df_final.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating_Count'} )
```

```python
#save the Group By output in a dataset which will be used for plotting the graphs
ratings= df_final.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating_Count'} )
ratings
```

# Observation Based on Ratings

1. 2.1k people has not given Ratings
2. Any Rating given below 2.4 and above 0 it is Poor Ratings
3. Rating between 2.5 to 3.4 is Average Ratings
4. Rating between 3.5 to 3.9, People have given Good Ratings
5. Rating between 4.0 to 4.4, People have given Very Good Rating
6. Rating between 4.5 to 4.9, People have given Excellent Ratings 

```python
#Plotting Graph based on Rating Count and Aggregate Rating
sns.set(rc = {'figure.figsize':(15,8)})
sns.barplot(x="Aggregate rating", y="Rating_Count", data=ratings,)
```

```python
#Plotting Graph based on Rating Count and Aggregate Rating and color of barplot is based on Rating Color
sns.set(rc = {'figure.figsize':(15,8)})
sns.barplot(x="Aggregate rating", y="Rating_Count", data=ratings,hue="Rating color", palette=['White','Red','Orange','Yellow','Green','Green',])
```

# Observations
1. Not Rated Count is Very High
2. Maximum number of Rating are between 2.5 and 3.4
